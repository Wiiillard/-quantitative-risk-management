## Problem 1
Given the dataset in DailyPrices.csv, for the stocks SPY, AAPL, and EQIX
1. Calculate the Arithmetic Returns. Remove the mean, such that each series has 0 mean.
Present the last 5 rows and the total standard deviation.
2. Calculate the Log Returns. Remove the mean, such that each series has 0 mean.
Present the last 5 rows and the total standard deviation.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("DailyPrices.csv")
data.head()
```

```python
stocks = ["SPY", "AAPL", "EQIX"]
prices = data[["Date"] + stocks]
prices
```

```python
arithmetic_returns = prices[stocks].pct_change().dropna()
arithmetic_returns
```

```python
arithmetic_returns_zero_mean = arithmetic_returns - arithmetic_returns.mean()
arithmetic_returns_zero_mean
```

```python
arithmetic_std = arithmetic_returns_zero_mean.std()
print("Arithmetic Returns Standard deviationï¼š")
arithmetic_std
```

```python
print("Arithmetic Returns of Last 5 Rows (Remove mean):")
print(arithmetic_returns_zero_mean.tail())
```

```python
log_returns = np.log(prices[stocks] / prices[stocks].shift(1)).dropna()
```

```python
log_returns_zero_mean = log_returns - log_returns.mean()
log_std = log_returns_zero_mean.std()
```

```python
print("Log Returns of Last 5 Rows(Remove Mean):")
print(log_returns_zero_mean.tail())
```

```python
print("Log Returns Standard deviationï¼š")
print(log_std)
```
